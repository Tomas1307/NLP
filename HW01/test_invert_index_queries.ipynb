```python

def read_queries(directory: str = "data/queries-raw-texts/"):
    # Ruta al directorio con los archivos .naf

    # Buscar todos los archivos que coincidan con el patrón
    files = glob.glob(directory + "wes2015.q*.naf")

    # Lista para almacenar los datos
    data = []

    # Procesar cada archivo
    for file in files:
        # Extraer el identificador del nombre del archivo
        identifier = file.split(".")[-2][-3:]  # Extrae '001', '002', etc.
        
        # Crear el objeto KafNafParser para el archivo actual
        naf_parser = KafNafParser(file)
        raw_text = naf_parser.get_raw()
        data.append({"identifier": identifier, "query": raw_text})
    
    return pd.DataFrame(data)

queries_df = read_queries()
queries_df
```

```python

def read_files(directory: str = "data/docs-raw-texts/"):
    # Ruta al directorio con los archivos .naf

    # Buscar todos los archivos que coincidan con el patrón
    files = glob.glob(directory + "wes2015.d*.naf")

    # Lista para almacenar los datos
    data = []

    # Procesar cada archivo
    for file in files:
        # Extraer el identificador del nombre del archivo
        identifier = file.split(".")[-2][-3:]  # Extrae '001', '002', etc.
        
        # Crear el objeto KafNafParser para el archivo actual
        naf_parser = KafNafParser(file)
        
        # Extraer el texto crudo
        raw_text = naf_parser.get_raw()

        title = naf_parser.root.find('nafHeader/fileDesc').get('title')
        
        # Agregar los datos a la lista
        data.append({"identifier": identifier, "text": raw_text, "title":title})

    df = pd.DataFrame(data)
    return df

df = read_files()


def replace_title_on_text(dataFrame):
    for i in range(0,len(dataFrame["text"])):
        if dataFrame["title"][i] in dataFrame["text"][i]:
            dataFrame["text"][i] = dataFrame["text"][i].replace(dataFrame["title"][i]+".","")
        if "\n" in dataFrame["text"][i]:
            dataFrame["text"][i] = dataFrame["text"][i].replace("\n","")


replace_title_on_text(df)



processor_ = Processor()

df["text"] = df["text"]+ " " + df["title"]
df.drop(columns=["title"],inplace=True)
for i in range(0,len(df["text"])):
    df["text"][i] = word_tokenize(df["text"][i])
    df["text"][i] = processor_.to_lowercase(df["text"][i])
    df["text"][i] = processor_.to_lowercase(df["text"][i])
    df["text"][i] = processor_.remove_punctuation(df["text"][i])
    df["text"][i] = processor_.remove_non_ascii(df["text"][i])
    df["text"][i] = processor_.remove_stopwords(df["text"][i])
    df["text"][i] = processor_.lemmatize_verbs(df["text"][i])
    df["text"][i] = ' '.join(df["text"][i])


df["text_list"] = [0] * 331
for i in range(0,len(df["text"])):
    df["text_list"][i] = df["text"][i].split()
df
```

```python
# Primero, extraemos las claves de inverted_index
inverted_index_keys = set(inverted_index.keys())

# Inicializamos listas para guardar los términos faltantes
missing_terms_df = []
missing_terms_queries_df = []

# Buscar términos faltantes en df["text"]
for idx, row in df.iterrows():
    text_terms = set(row['text_list'])
    missing_in_text = text_terms - inverted_index_keys
    if missing_in_text:
        missing_terms_df.append({'identifier': row['identifier'], 'missing_terms': list(missing_in_text)})

# Buscar términos faltantes en queries_df["query"]
for idx, row in queries_df.iterrows():
    query_terms = set(row['query_list'])
    missing_in_query = query_terms - inverted_index_keys
    if missing_in_query:
        missing_terms_queries_df.append({'identifier': row['identifier'], 'missing_terms': list(missing_in_query)})

# Convertimos los resultados a DataFrames para visualización
missing_terms_df = pd.DataFrame(missing_terms_df)
missing_terms_queries_df = pd.DataFrame(missing_terms_queries_df)

print("Términos faltantes en df['text']:")
print(missing_terms_df)

print("\nTérminos faltantes en queries_df['query']:")
print(missing_terms_queries_df)
```
