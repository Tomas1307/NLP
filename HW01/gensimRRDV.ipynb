```python
def sorted_query(lista):
    resultado = []
    
    for item in lista:
        query, documentos_str = item.split(' ', 1)
        documentos = documentos_str.split(',')
        
        documentos_ordenados = sorted(documentos, key=lambda x: float(x.split(':')[1]), reverse=True)
        
        resultado.append(f"{query} {','.join(documentos_ordenados)}")
    
    return resultado

def format_cosine_similarities(df_queries_small, df_texts_small, RRDV_):
    results = {}

    for i in range(len(df_queries_small)):
        query_id = df_queries_small['identifier'].iloc[i]
        query_results = []

        for j in range(len(df_texts_small)):
            value = RRDV_.cosine_similarity(df_queries_small["query_list"].iloc[i], df_texts_small["text_list"].iloc[j])
            if value > 0:
                doc_id = f"d{df_texts_small['identifier'].iloc[j]}"
                query_results.append(f"{doc_id}:{value:.6f}")

        results[query_id] = ','.join(query_results)

    formatted_output = []
    for query_id, result in results.items():
        formatted_output.append(f"{query_id} {result}")
        
    
    #formatted_output = sorted_query(formatted_output)
    
    print(formatted_output)

    return '\n'.join(formatted_output)

```

```python
formatted_result = format_cosine_similarities(df_queries_small, df_texts_small, RRDV_)

formatted_result
```
