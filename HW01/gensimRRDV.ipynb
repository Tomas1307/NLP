```python
with open("inverted_index.json") as file:
    inverted_index_file = json.load(file)
    
inverted_index_file
```

```python
documentos = {}
for termino, docs in inverted_index_file.items():
    for doc_id in docs:
        if doc_id not in documentos:
            documentos[doc_id] = []
        documentos[doc_id].append(termino)

# Convertir a lista de documentos (cada documento es una lista de trminos)
documentos_list = [terminos for doc_id, terminos in sorted(documentos.items())]
documentos_list
```

```python
dictionary = corpora.Dictionary(documentos_list)
corpus = [dictionary.doc2bow(doc) for doc in documentos_list]
```
